```python
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

## Baseline Model
Milestone 1 - Group 4 - Turtle Forcast:

Value of Product:
Predict the number of turtles caught in 2019 per week and capture site.

Prediction:
Its possible to predict the number of turtles from the past.

Evaluation Metric:
RMSE, R^2
We want to be as close to the real number as possible.

Baseline Model:
We think that taking the number from the previous year will be a good prediction for the current year.

Score:
RMSE = 2.9, R^2 = -0.7

## Load Data
This should later be in a py function...

```python
turtles = pd.read_csv('data/train.csv')
```

```python
turtles_minimal = turtles[['Date_TimeCaught', 'CaptureSite']]
```

```python
# Make a Year and a Calender Week Column
turtles_minimal.Date_TimeCaught = pd.to_datetime(turtles_minimal.Date_TimeCaught)
turtles_minimal['CalendarWeek'] = turtles_minimal['Date_TimeCaught'].dt.isocalendar().week
turtles_minimal['Year'] = turtles_minimal['Date_TimeCaught'].dt.year
```

```python
# Training Data for 2017
turtles_2017 = turtles_minimal[turtles_minimal.Year == 2017]
counts_per_week_site_2017 = turtles_2017.groupby(['CalendarWeek', 'CaptureSite']).size().reset_index(name='Count')

# Test Data for 2018
turtles_2018 = turtles_minimal[turtles_minimal.Year == 2018]
counts_per_week_site_2018 = turtles_2018.groupby(['CalendarWeek', 'CaptureSite']).size().reset_index(name='Count')
```

```python
# Make the "prediction"
counts_per_week_site_2018 = counts_per_week_site_2018.merge(
    counts_per_week_site_2017[['CalendarWeek', 'CaptureSite', 'Count']],
    on=['CalendarWeek', 'CaptureSite'],
    how='left',
    suffixes=('_2018', '_2017')
)
```

```python
# Ahhh. there are some missing values now...
counts_per_week_site_2018.fillna(0, inplace=True)
```

```python
#calculate RMSE for Prediction
realVals = list(counts_per_week_site_2018.Count_2018)
predictedVals = list(counts_per_week_site_2018.Count_2017)
mse = mean_squared_error(realVals, predictedVals)
rmse = mean_squared_error(realVals, predictedVals, squared = False)
R2Score = r2_score(realVals, predictedVals)
```
